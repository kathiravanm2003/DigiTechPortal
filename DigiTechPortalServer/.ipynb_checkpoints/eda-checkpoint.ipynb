```python
# 1. Loading the data
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data from a CSV file
sales_data = pd.read_csv('./content/sales2.csv')

print(sales_data)
```

```python
# 2. Data cleaning and pre-processing:
# Handling missing values
sales_data = sales_data.dropna()

# Dealing with outliers
sales_data = sales_data[(sales_data['Sales'] > 0) & (sales_data['Sales'] < 5000)]

# Transforming the data
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data['Year'] = sales_data['Date'].dt.year
sales_data['Month'] = sales_data['Date'].dt.month

# sales_data['new_date'] = sales_data['Date'] + '/' sales_data['Month'] + '/' + sales_data['Year']

print(sales_data)
# print(sales_data['new_date'])
```

```python
# Data visualization:

# Create a histogram of the sales data
plt.hist(sales_data['Sales'], bins=20)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()

# Create a line chart of sales by month
print(sales_data)
sales_by_month = sales_data.groupby('Month')['Sales'].sum()
print(sales_by_month)
plt.plot(sales_by_month)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
# 4. Data analysis:
# Descriptive statistics
print(sales_data['Sales'].describe())
print('=============================')

# Correlation analysis
print(sales_data[['Sales', 'Profit']].corr())
print('=============================')

# Hypothesis testing
from scipy.stats import ttest_ind

# Compare sales by region
north_sales = sales_data[sales_data['Region'] == 'North']['Sales']
south_sales = sales_data[sales_data['Region'] == 'South']['Sales']

if north_sales.std() != 0 and south_sales.std() != 0:
    t_stat, p_val = ttest_ind(north_sales, south_sales)
    print("t-statistic: ", t_stat)
    print("p-value: ", p_val)
else:
    print("One of the standard deviations is zero, cannot perform t-test.")
print('=============================')
print(north_sales)
print('=============================')
print(south_sales)
print('=============================')
print(ttest_ind(north_sales, south_sales))

"""
The ttest_ind function takes two arrays of data as input and returns the calculated t-statistic and p-value for the test.
The t-statistic is a measure of the difference between the means of the two samples, while the p-value 
indicates the probability of observing the difference in means by chance if there is no real difference between the two populations.
"""
```

```python
# 5. Insights and conclusions:
print("""
 Based on the EDA, we can draw the following insights and conclusions:
 1.The distribution of sales is right-skewed and most of the sales are between 0 and 2000.
 2.Sales are highest in the months of November and December.
 3.There is a strong positive correlation between sales and profit.
 4.Sales in the North region are significantly higher than sales in the South region.""")
```

```python
# 6. Sales Prediction

# Load data from CSV file
sales_data = pd.read_csv('./content/sales2.csv')

# Convert date string to datetime object and extract month
sales_data['Month'] = pd.to_datetime(sales_data['Date']).dt.month

# Group sales data by month and calculate mean sales
monthly_sales = sales_data.groupby('Month')['Sales'].mean().reset_index()

# Create a LinearRegression object
lr = LinearRegression()

# Train the model on the monthly sales data
print(monthly_sales)
lr.fit(monthly_sales[['Month']], monthly_sales['Sales'])

# Predict sales for each month of year 2021
year_2021 = pd.DataFrame({'Month': range(1, 13)})
predicted_sales = lr.predict(year_2021)

# Print the predicted sales for each month of year 2021
print('Predicted Sales Q4 for Year 2021:')
print(predicted_sales)

# Plot the predicted sales for each month of year 2021
plt.plot(year_2021['Month'], predicted_sales, label='Predicted Sales')

# Plot the monthly sales data for comparison
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], label='Monthly Sales')

# Add title and labels to the plot
plt.title('Monthly Sales for Q4 Year 2021')
plt.xlabel('Month')
plt.ylabel('Sales')

# Add legend to the plot
plt.legend()

# Show the plot
plt.show()
```
