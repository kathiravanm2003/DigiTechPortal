```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
data=pd.read_csv(r'C:\Users\selven\source\repos\DigiTechPortal\DigiTechPortalServer\content\sales_data.csv')   

# Split the data into training and testing sets
train_data = data[:-12]
test_data = data[-12:]

# Create the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(train_data[['actual_sales']], train_data['predicted_sales'])

# Predict the sales on the testing data
predictions = model.predict(test_data[['actual_sales']])

# Visualize the results
plt.plot(test_data['date'], test_data['predicted_sales'], label='Predicted Sales')
plt.plot(test_data['date'], test_data['actual_sales'], label='Actual Sales')

plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the data
data=pd.read_csv(r'C:\Users\selven\source\repos\DigiTechPortal\DigiTechPortalServer\content\sales_data.csv')   

# Prompt the user for the date range
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Filter the data by date range
filtered_data = data[(data['date'] >= start_date) & (data['date'] <= end_date)]

# Create the linear regression model
model = LinearRegression()

# Train the model on the filtered data
model.fit(filtered_data[['actual_sales']], filtered_data['predicted_sales'])

# Predict the sales for the specified time period
sales_data = pd.DataFrame({'actual_sales': [0, 0, 0]}) # Create a DataFrame with three rows for the three months
predictions = model.predict(sales_data)

# Print the predicted sales for the specified time period
print("Predicted sales for the period from", start_date, "to", end_date)
for i, prediction in enumerate(predictions):
    month = pd.to_datetime(end_date) + pd.DateOffset(months=i+1)
    print(month.strftime('%B %Y'), "-", round(prediction, 2))
```
